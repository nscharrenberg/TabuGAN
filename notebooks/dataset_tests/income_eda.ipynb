# Exploratory Data Analysis for Bank Marketing Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_path = '../../datasets/income/adult.csv'
df = pd.read_csv(dataset_path)
```

```python
df.head()
```

```python
categorical_columns = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country', 'income']
```

```python
df.describe().T
```

```python
# Set up the plot grid
fig, axs = plt.subplots(3, 3, figsize=(20, 15))

# List of categorical columns and corresponding axes
columns_axes = zip(categorical_columns, axs.flatten())

# Plot the distributions
for col, ax in columns_axes:
    df[col].value_counts(normalize=True).plot(kind='bar', ax=ax, title=col, rot=45)
    ax.set_ylabel('Percentage')
    ax.set_xlabel(col)

# Adjust layout for better visualization
plt.tight_layout()

# Display the plots
plt.show()
```
