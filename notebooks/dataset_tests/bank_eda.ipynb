# Exploratory Data Analysis for Bank Marketing Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_path = '../../datasets/bank/bank.csv'
df = pd.read_csv(dataset_path)
```

```python
df.head()
```

Define the columns that we want to explore, in our case we want to check if there is a class imbalance within our dataset.

```python
categorical_columns = ['deposit', 'job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']
```

```python
df.describe().T
```

Let's plot each columns distribution, so we can evaluate the dataset a bit better.

```python
# Set up the plot grid
fig, axs = plt.subplots(4, 3, figsize=(20, 15))

# List of categorical columns and corresponding axes
columns_axes = zip(categorical_columns, axs.flatten())

# Plot the distributions
for col, ax in columns_axes:
    df[col].value_counts(normalize=True).plot(kind='bar', ax=ax, title=col, rot=45)
    ax.set_ylabel('Percentage')
    ax.set_xlabel(col)

# Adjust layout for better visualization
plt.tight_layout()

# Display the plots
plt.show()
```

The target variable "deposit" is relatively balanced, but other features exhibit notable imbalances that could affect diversity metrics, namely:
- **Heavy**:
    - Default
    - Loan
    - Contact
    - poutcome
- **Slight**:
    - Job
    - Marital
    - Education
    - Month
